```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def gauss(x, mu, sigma): 
    return np.exp(-(x - mu)**2/(2*sigma**2))/(np.sqrt(2*np.pi)*sigma) #Se define una función para calcular la densidad
                                                                      #de probabilidad gaussina 
```

```python
x1=np.linspace(np.min(c1),np.max(c1),100) #se definen los valores x en la cual será evaluada la función gauss
x2=np.linspace(np.min(c2),np.max(c2),100) #se utiliza como rango el máximo y el mínimo de cada conjunto.
x3=np.linspace(np.min(c3),np.max(c3),100)
```

Con el fin de estudiar los conjuntos se procede a graficar cada uno de ellos en un histograma con 10 intervalos y también la curva correspondiente a la función de distribución de probabilidad gaussiana asociada.\
Recordar que el intervalo:
$$
[\mu-\sigma,\mu+\sigma]
$$
\
Contiene el 68% de los datos en una distribución gaussiana. Este intervalo se representa con una lineas punteadas verticales en los gráficos.\
**Observar que para poder comparar la forma del histograma con la función de probabilidad se debe "normalizar" el histograma, por lo tanto a pesar de utilizar el código de histograma, lo que realmente se obtiene es un histograma con frecuencias relativas.**


```python
fig=plt.figure(figsize=(5,12)) #Crear la figura y 3 subplots
ax1 = fig.add_subplot(3,1,1)
ax2 = fig.add_subplot(3,1,2)
ax3 = fig.add_subplot(3,1,3)

ax1.hist(c1, color='SteelBlue', alpha=1, edgecolor = 'white',  linewidth=1, bins=10, density=True) #Se grafican todos los 
ax1.plot(x1, gauss(x1, 10, 0.1), color='black')                                                    #histogramas normalizados
ax1.axvline(x=10-0.1, ymin=0, ymax=0.56,color='red',linestyle='--')                                #y sus respectivas funciones
ax1.axvline(x=10+0.1, ymin=0, ymax=0.56,color='red',linestyle='--')                                #de probabilidad
 
ax2.hist(c2, color='SteelBlue', alpha=1, edgecolor = 'white',  linewidth=1, bins=10, density=True)
ax2.plot(x2, gauss(x2, 10, 0.1),color='black')
ax2.axvline(x=10-0.1, ymin=0, ymax=0.56,color='red',linestyle='--')
ax2.axvline(x=10+0.1, ymin=0, ymax=0.56,color='red',linestyle='--')

ax3.hist(c3, color='SteelBlue', alpha=1, edgecolor = 'white',  linewidth=1, bins=10, density=True)
ax3.plot(x3, gauss(x3, 10, 0.1),color='black')
ax3.axvline(x=10-0.1, ymin=0, ymax=0.29,color='red',linestyle='--')
ax3.axvline(x=10+0.1, ymin=0, ymax=0.29,color='red',linestyle='--')

fig.suptitle('Distribución de datos en los conjuntos',size=15)#Título general
ax1.set_title('Conjunto 1')#Subtitulos de cada grafico
ax2.set_title('Conjunto 2')
ax3.set_title('Conjunto 3')
fig.tight_layout(pad=1.0)

ax1.set_xlabel('Mediciones')
ax1.set_ylabel('Frecuencia Relativa')
ax2.set_xlabel('Mediciones')
ax2.set_ylabel('Frecuencia Relativa')
ax3.set_xlabel('Mediciones')
ax3.set_ylabel('Frecuencia Relativa')
```

De lo anterior se puede apreciar que una distribución gaussiana se ajusta bastante bien con los datos del **conjunto 2**. Se puede apreciar también que en el **conjunto 3** todos los datos se encuentran en el intervalo [𝜇−𝜎,𝜇+𝜎], es decir, los datos son más homogéneos y no obedecen una distribución gaussiana. Por último se puede apreciar el **conjunto 1** posee unos valores aberrantes que se alejan mas de 3𝜎 del valor medio real. Por lo tanto se puede realizar el ejercicio de volver a ajustar una función de probabilidad gaussiana, omitiendo estos datos aberrantes.

```python
c1a=np.array(c1) #se convierte la lista del conjunto 2 en un arreglo
```

```python
c1n=np.delete(c1a,c1a>12) #se crea un conjunto 2 nuevo, en el cual se omite todo valor mayor a 12
```

```python
x1n=np.linspace(np.min(c1n),np.max(c1n),100) # se definen nuevos valores x para evaluar la función gauss
```

```python
fig=plt.figure()
ax = fig.add_subplot(1,1,1)
ax.hist(c1n, color='SteelBlue', alpha=1, edgecolor = 'white',  linewidth=1, bins=10, density=True)
ax.plot(x1n, gauss(x1n, 10, 0.1), color='black')
ax.axvline(x=10-0.1, ymin=0, ymax=0.56,color='red',linestyle='--')
ax.axvline(x=10+0.1, ymin=0, ymax=0.56,color='red',linestyle='--')
ax.set_title('Conjunto 1 nuevo',size=15)
ax.set_xlabel('Mediciones')
ax.set_ylabel('Frecuencia Relativa')
fig.tight_layout(pad=1.0)
```

Se puede apreciar que omitiendo los valores aberrantes el conjunto 1 se ajusta con bastante precisión con la función de distribución gaussiana.
